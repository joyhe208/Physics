```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
```

```python
M = 1
l = 1
g = 9.8
```

```python
#Osciallation period with small angle approximation:
T1 = 2*math.pi*math.sqrt(l/g)
T1
```

```python
#plotting using integration
v0 = 3
times = np.arange(0, 10.01, .01)
v = []
a = []
theta = []
times
step = .01
for i in range(len(times)):
    if(i == 0):
        a.append(0)
        v.append(v0)
        theta.append(0)
    else:
        theta.append(theta[i-1] + l*v[i-1]*step)
        a.append(g*math.sin(theta[i]))
        v.append(v[i-1] - a[i]*step)
theta2 = []
theta0 = math.sqrt(9/(l*g))
for i in range(len(times)):
    theta2.append(theta0*math.sin(math.sqrt(g/l)*times[i]))
fig = plt.figure(figsize=(10,6))
Plot = fig.add_subplot(111)
#Plot.plot(times,theta)
Plot.plot(times, theta2)
```
