```python
import numpy as np
import math
import matplotlib.pyplot as plt
from numericalunits import *
import pint
ureg = pint.UnitRegistry()
```

```python
w = 3
a0 = -1
#calculating displacement using antiderivatives
x = a0*((-1/9)*math.sin(2*math.pi*w) + (1/3)*2*math.pi)
print("actual displacement:", x, "cm")
```

```python
times = np.arange(0,10.1,.1)
velocities = []
displacements = []

for i in range(len(times)):
    if(i ==0):
        velocities.append(0)
    else:
        velocities.append(-math.sin(3*times[i-1])*.1 + velocities[i-1])
        
for i in range(len(times)):
    if(i==0):
        displacements.append(0)
    else:
        displacements.append(velocities[i-1]*.1 + displacements[i-1])
        
fig = plt.figure(figsize=(10,6))
Plot = fig.add_subplot(111)
Plot.plot(times,velocities)
Plot.plot(times, displacements)
print("displacement calculated by approximation:", displacements[63], "cm")
print("error (difference between b and a):" , displacements[63]-x, "cm")
```

```python
times2 = np.arange(-8,10.1,.1)
accelerations = []
for time in times2:
    if time<0:
        accelerations.append(-1*(10**(.5*time)/(.01+10**(.5*time)))*math.sin(3*time))
    else:
        accelerations.append(-1*math.sin(3*time))
        
velocities2 = []
displacements2 = []

for i in range(len(times2)):
    if(i ==0):
        velocities2.append(0)
    else:
        velocities2.append(accelerations[i-1]*.1 + velocities2[i-1])
        
for i in range(len(times2)):
    if(i==0):
        displacements2.append(0)
    else:
        displacements2.append(velocities2[i-1]*.1+displacements2[i-1])
        
fig = plt.figure(figsize=(10,6))
Plot2 = fig.add_subplot(111)
Plot2.plot(times2,velocities2)
Plot2.plot(times2, displacements2)
print("displacement when the oscillating field is turned on slowly:", displacements2[143], "cm")
```

```python
print("difference between c and a:", displacements2[143]-x, "cm")
```
