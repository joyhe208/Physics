```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from IPython.display import Image
imagepath = '/Users/joyhe208/desktop/midtermprob.png'
Image(imagepath, width=600, height=200)
```

```python
M = 5
lamb = 2
k = 5
t = np.arange(0,30.01, .01)
x = []
D = 1
l = 1
for i in range(len(t)):
    x.append((D)*math.cos(math.sqrt(k/M)*t[i])+l)
def solvr(Y,t):
    return[Y[1],(-k/M)*Y[0]-(lamb/M)*Y[1]];
x2 = odeint(solvr, [D,0],t)[:,0]
x3 = []
w1 = math.sqrt((k/M)-(lamb/(2*M))**2)
for time in t:
    x3.append(D*math.exp((-lamb*time)/(2*M))*math.cos(w1*time)+l)
fig = plt.figure(figsize=(10,6))
Plot = fig.add_subplot(111)
Plot.plot(t, x)
Plot.plot(t, x3)
```
